```python
"""
When we assume that no part overlaps another,
on exchanging the top left and the bottom left parts,
the bottom left part of the final image does not align with the rest of the image

If we overlap a bit of upper left part by the bottom left part (so that both the upper
left and the bottom left parts have the same dimensions (200 pixels x 190 pixels) ),
the parts align just fine, but a gap of dimensions (10 pixels x 190 pixels) remains
at the bottom left of the final image
"""
```

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
image = Image.open("/content/jigsaw.jpg")

image_1=np.asarray(image)
```

```python
raw_data_1 = image_1.copy()

# reflecting the chick in the midddle laterally
for i in range(150,330) :
    for j in range(515,700) :
        for x in range(3) :
            raw_data_1[i][j][x]=image_1[i][700+515-j-1][x]
```

```python
raw_data_2 = raw_data_1.copy()

# reflecting the part at the bottom right vertically
for i in range(369,421) :
    for j in range(370,797) :
        for x in range(3) :
            raw_data_2[i][j][x]=raw_data_1[369+421-i-1][j][x]
```

```python
raw_data_3=raw_data_2.copy()

# reflecting vertically the part at the bottom left
# exchanging parts originally present at top left and bottom left

for i in range(0,210) :
    for j in range(0,190) :
        for x in range(3) :
            raw_data_3[i][j][x]=raw_data_2[409-i][j][x]

raw_data_4=raw_data_3.copy()

for i in range(210,410) :
    for j in range(0,190) :
        for x in range(3) :
            raw_data_4[i][j][x]=raw_data_2[i-210][j][x]
```

```python
io.imshow(raw_data_4)
```

```python
plt.imshow(raw_data_4)
```
